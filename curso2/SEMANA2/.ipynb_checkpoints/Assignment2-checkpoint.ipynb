```python
import matplotlib.pyplot as plt
import mplleaflet
import pandas as pd

def leaflet_plot_stations(binsize, hashid):

    df = pd.read_csv('data/C2A2_data/BinnedCsvs_d400/BinSize_d{}.csv'.format(binsize))
    
    station_locations_by_hash = df[df['hash'] == hashid]

    lons = station_locations_by_hash['LONGITUDE'].tolist()
    lats = station_locations_by_hash['LATITUDE'].tolist()

    plt.figure(figsize=(8,8))

    plt.scatter(lons, lats, c='r', alpha=0.7, s=200)

    return mplleaflet.display()

leaflet_plot_stations(400,'fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89')
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as dates
import numpy as np
```

```python
df=pd.read_csv('data/C2A2_data/BinnedCsvs_d400/fb441e62df2d58994928907a91895ec62c2c42e6cd075c2700843b89.csv')
```

```python
# extract day of year from datetime object
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values("Date")
#Data_Value column by 10 to get Values in degree c
df["Data_Value"] = df["Data_Value"]/10

#convertir object to datetime 
df['Date']=pd.to_datetime(df['Date'])

#extract day and month of year from datetime object
df['M-D']=df['Date'].dt.strftime('%m-%d')

#Watch out for leap days (i.e. February 29th),
#it is reasonable to remove these points from the dataset for the purpose of this visualization.
df.drop(df[df['Date']=='2008-02-29'].index, inplace=True)
df.drop(df[df['Date']=='2012-02-29'].index, inplace=True)



# extract the max of max and min of min tempratures
max_temp=df[(df['Date'] >= '2005') & (df['Date'] < '2015')
            & (df['Element'] == 'TMAX')].groupby(['M-D'])['Data_Value'].max()
```

```python
max_temp.reset_index()
```

```python
min_temp=df[(df['Date'] >= '2005') & (df['Date'] < '2015') & 
            (df['Element'] == 'TMIN')].groupby(['M-D'])['Data_Value'].min()
```

```python
df = df.merge(max_temp.reset_index(False).rename(columns={'Data_Value':'Max_temp'}), on='M-D', how='left')
```

```python
df = df.merge(min_temp.reset_index(drop=False).rename(columns={'Data_Value':'Min_temp'}), on='M-D', how='left')
```

```python
record_high = df[(df['Date'] >= '2015')&(df.Data_Value > df.Max_temp)]
record_high.head(2) 
```

```python
record_low = df[(df['Date'] >= '2015')&(df.Data_Value < df.Min_temp)]
```
